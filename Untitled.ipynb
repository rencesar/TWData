```python
import pandas as pd
import numpy as np

from selenium import webdriver
from selenium.webdriver.chrome.options import Options

from bs4 import BeautifulSoup
```

```python
options = Options()
options.add_argument("--user-data-dir=/Users/renato/Library/Application Support/Google/Chrome/")
options.add_argument(r'--profile-directory=Profile 2') #e.g. Profile 3
driver = webdriver.Chrome(executable_path=r'/Users/renato/Downloads/chromedriver', chrome_options=options)
driver.get("https://br108.tribalwars.com.br/page/play/br108")
driver.get("https://br108.tribalwars.com.br/game.php?screen=ally&mode=members_troops")
```

```python
url_trops_players = "https://br108.tribalwars.com.br/game.php?screen=ally&mode=members_troops&player_id={player_id}"
html_members_selection = driver.page_source 
soup = BeautifulSoup(html_members_selection, 'html.parser')
players = [[opt['value'], opt.text] for opt in soup.select('select[name=player_id] > option[value]')]
```

```python
result_list = []
for player_id, name in players:
    url_trop_player = url_trops_players.format(player_id=player_id)
    driver.get(url_trop_player)
    html_player_trops = driver.page_source 
    soup = BeautifulSoup(html_player_trops, 'html.parser')
    table = soup.find('table', {"class": "vis w100"})
    aldeias = table.find_all('tr')[1:]
    total = []
    for aldeia in aldeias:
        result_aldeia = [tr.text.strip() for tr in aldeia.find_all('td')]
        result_aldeia.insert(0, name)
        total.append(result_aldeia)

    result_list += total
```

```python
df = pd.DataFrame(result_list)
```

```python
df.columns = ['Player', 'Aldeia', 'Lança', 'Espada', 'Machado', 'Explorador', 'Cavalaria Leve', 'Cavalaria Pesada', 'Ariete', 'Cataputa', 'Paladino', 'Nobre', 'Milicia', 'Comandos ativos', 'Ataques a chegar']
```

```python
df = pd.read_excel('tropas_tribo.xlsx')
```

```python
df.convert_dtypes().dtypes
```

```python
df['Lança'] = pd.to_numeric(df['Lança'], errors='coerce')
df['Espada'] = pd.to_numeric(df['Espada'], errors='coerce')
df['Machado'] = pd.to_numeric(df['Machado'], errors='coerce')
df['Explorador'] = pd.to_numeric(df['Explorador'], errors='coerce')
df['Cavalaria Leve'] = pd.to_numeric(df['Cavalaria Leve'], errors='coerce')
df['Cavalaria Pesada'] = pd.to_numeric(df['Cavalaria Pesada'], errors='coerce')
df['Ariete'] = pd.to_numeric(df['Ariete'], errors='coerce')
df['Cataputa'] = pd.to_numeric(df['Cataputa'], errors='coerce')
df['Paladino'] = pd.to_numeric(df['Paladino'], errors='coerce')
df['Nobre'] = pd.to_numeric(df['Nobre'], errors='coerce')
df['Milicia'] = pd.to_numeric(df['Milicia'], errors='coerce')

df = df.fillna(0)
```

```python
df.convert_dtypes().dtypes
```

```python
df['Total Tropas Defesa'] = df['Lança'] + df['Espada'] + df['Cavalaria Pesada']
```

```python
df['Total Tropas Ataque'] = df['Machado'] + df['Cavalaria Leve'] + df['Ariete']
```

```python
df['Total Tropas'] = df['Total Tropas Defesa'] + df['Total Tropas Ataque'] + df['Cataputa']
```

```python
df['Porcentagem Tropas Ataque'] = (df['Total Tropas Ataque'] / df['Total Tropas']) * 100
```

```python
df['Porcentagem Tropas Defesa'] = (df['Total Tropas Defesa'] / df['Total Tropas']) * 100
```

```python
df['Tipo Aldeia'] = pd.cut(df['Porcentagem Tropas Ataque'], 3, labels=['Defesa', 'Misto', 'Ataque'])
```

```python
labels = ["Poucas", "Meio Full", "Quase Full", "Full"]

bins_bb = pd.IntervalIndex.from_tuples([(999, 3000), (3000, 5000), (5000, 7000), (7000, np.inf)])
bb_qtd = pd.cut(df['Machado'].values, bins_bb)
bb_qtd= bb_qtd.rename_categories(labels)
df['BB Qtd'] = bb_qtd

bins_cl = pd.IntervalIndex.from_tuples([(0, 500), (500, 1500), (1500, 2500), (2500, np.inf)])
cl_qtd = pd.cut(df['Cavalaria Leve'].values, bins_cl)
cl_qtd= cl_qtd.rename_categories(labels)
df['CL Qtd'] = cl_qtd

bins_cl = pd.IntervalIndex.from_tuples([(0, 100), (100, 200), (200, 299), (299, np.inf)])
cl_qtd = pd.cut(df['Ariete'].values, bins_cl)
cl_qtd= cl_qtd.rename_categories(labels)
df['Ariete Qtd'] = cl_qtd
```

```python
df_ataque = df[df['Tipo Aldeia'] == 'Ataque']
```

```python
text_to_forum = ''
current_player = ''
for index, row in df_ataque.iterrows():
    coords_aldeia = row['Aldeia'][row['Aldeia'].find("(")+1:row['Aldeia'].find(")")]
    if current_player != row['Player']:
        current_player = row['Player']
        text_to_forum += f"""[/table]\n\n\n
Player: [player]{row['Player']}[/player]


[table]
[**]Aldeia[building]main[/building][||][unit]axe[/unit][||][unit]spy[/unit][||][unit]light[/unit][||][unit]ram[/unit][||][unit]catapult[/unit][||][unit]knight[/unit][||][unit]snob[/unit][||]BB Status[||]CL Status[||]Ariete Status[||][/**]
"""
    text_to_forum += f"[*][coord]{coords_aldeia}[/coord] [|]{row['Machado']} [|] {row['Explorador']} [|] {row['Cavalaria Leve']} [|] {row['Ariete']} [|] {row['Cataputa']} [|] {row['Paladino']} [|] {row['Nobre']} [|] {row['BB Qtd']} [|] {row['CL Qtd']} [|] {row['Ariete Qtd']} \n"
text_to_forum += '[/table]'
print(text_to_forum)
```

```python
df_ataque.groupby('Player').agg({'BB Qtd': 'count',
                          'CL Qtd':'count',
                          'Ariete Qtd':'count',
                          'Nobre':'sum'})
```

```python
df_ataque.apply(lambda x: pd.Series({
    'Full BB':(x['BB Qtd'] == 'Full').sum(),
    'Quase Full BB':(x['BB Qtd'] == 'Quase Full').sum(),
    'Meio Full BB':(x['BB Qtd'] == 'Meio Full').sum(),
    'Poucas BB':(x['BB Qtd'] == 'Poucas').sum(),
    'Full CL':(x['CL Qtd'] == 'Full').sum(),
    'Quase Full CL':(x['CL Qtd'] == 'Quase Full').sum(),
    'Meio Full CL':(x['CL Qtd'] == 'Meio Full').sum(),
    'Poucas CL':(x['CL Qtd'] == 'Poucas').sum(),
    'Full Ariete':(x['Ariete Qtd'] == 'Full').sum(),
    'Quase Full Ariete':(x['Ariete Qtd'] == 'Quase Full').sum(),
    'Meio Full Ariete':(x['Ariete Qtd'] == 'Meio Full').sum(),
    'Poucas Ariete':(x['Ariete Qtd'] == 'Poucas').sum(),
}))
```

```python
bb_full = (df_ataque['BB Qtd'] == 'Full').groupby(df_ataque['Player']).sum().rename('BB Full')
bb_q_full = (df_ataque['BB Qtd'] == 'Quase Full').groupby(df_ataque['Player']).sum().rename('BB Quase Full')
bb_meio_full = (df_ataque['BB Qtd'] == 'Meio Full').groupby(df_ataque['Player']).sum().rename('BB Meio Full')
bb_poucas = (df_ataque['BB Qtd'] == 'Poucas').groupby(df_ataque['Player']).sum().rename('BB Poucas')

cl_full = (df_ataque['CL Qtd'] == 'Full').groupby(df_ataque['Player']).sum().rename('CL Full')
cl_q_full = (df_ataque['CL Qtd'] == 'Quase Full').groupby(df_ataque['Player']).sum().rename('CL Quase Full')
cl_meio_full = (df_ataque['CL Qtd'] == 'Meio Full').groupby(df_ataque['Player']).sum().rename('CL Meio Full')
cl_poucas = (df_ataque['CL Qtd'] == 'Poucas').groupby(df_ataque['Player']).sum().rename('CL Poucas')

ariete_full = (df_ataque['Ariete Qtd'] == 'Full').groupby(df_ataque['Player']).sum().rename('Ariete Full')
ariete_q_full = (df_ataque['Ariete Qtd'] == 'Quase Full').groupby(df_ataque['Player']).sum().rename('Ariete Quase Full')
ariete_meio_full = (df_ataque['Ariete Qtd'] == 'Meio Full').groupby(df_ataque['Player']).sum().rename('Ariete Meio Full')
ariete_poucas = (df_ataque['Ariete Qtd'] == 'Poucas').groupby(df_ataque['Player']).sum().rename('Ariete Poucas')

data = [
    bb_full, bb_q_full, bb_meio_full, bb_poucas,
    cl_full, cl_q_full, cl_meio_full, cl_poucas,
    ariete_full, ariete_q_full, ariete_meio_full, ariete_poucas,
]

tropas_df = pd.concat(data, axis=1, keys=[s.name for s in data])
```
