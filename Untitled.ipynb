```python
import pandas as pd
import numpy as np

from selenium import webdriver
from selenium.webdriver.chrome.options import Options

from bs4 import BeautifulSoup
```

```python
options = Options()
options.add_argument("--user-data-dir=/Users/renato/Library/Application Support/Google/Chrome/")
options.add_argument(r'--profile-directory=Profile 2') #e.g. Profile 3
driver = webdriver.Chrome(executable_path=r'/Users/renato/Downloads/chromedriver', chrome_options=options)
driver.get("https://br108.tribalwars.com.br/page/play/br108")
driver.get("https://br108.tribalwars.com.br/game.php?screen=ally&mode=members_troops")
```

```python
url_trops_players = "https://br108.tribalwars.com.br/game.php?screen=ally&mode=members_troops&player_id={player_id}"
html_members_selection = driver.page_source 
soup = BeautifulSoup(html_members_selection, 'html.parser')
players = [[opt['value'], opt.text] for opt in soup.select('select[name=player_id] > option[value]')]
```

```python
result_list = []
for player_id, name in players:
    url_trop_player = url_trops_players.format(player_id=player_id)
    driver.get(url_trop_player)
    html_player_trops = driver.page_source 
    soup = BeautifulSoup(html_player_trops, 'html.parser')
    table = soup.find('table', {"class": "vis w100"})
    aldeias = table.find_all('tr')[1:]
    total = []
    for aldeia in aldeias:
        result_aldeia = [tr.text.strip() for tr in aldeia.find_all('td')]
        result_aldeia.insert(0, name)
        total.append(result_aldeia)

    result_list += total
```

```python
df = pd.DataFrame(result_list)
```

```python
df.columns = ['Player', 'Aldeia', 'Lança', 'Espada', 'Machado', 'Explorador', 'Cavalaria Leve', 'Cavalaria Pesada', 'Ariete', 'Cataputa', 'Paladino', 'Nobre', 'Milicia', 'Comandos ativos', 'Ataques a chegar']
```

```python
df['Player'] = df['Player'].apply(lambda x: x.strip())
```

```python
df['Lança'] = pd.to_numeric(df['Lança'], errors='coerce')
df['Espada'] = pd.to_numeric(df['Espada'], errors='coerce')
df['Machado'] = pd.to_numeric(df['Machado'], errors='coerce')
df['Explorador'] = pd.to_numeric(df['Explorador'], errors='coerce')
df['Cavalaria Leve'] = pd.to_numeric(df['Cavalaria Leve'], errors='coerce')
df['Cavalaria Pesada'] = pd.to_numeric(df['Cavalaria Pesada'], errors='coerce')
df['Ariete'] = pd.to_numeric(df['Ariete'], errors='coerce')
df['Cataputa'] = pd.to_numeric(df['Cataputa'], errors='coerce')
df['Paladino'] = pd.to_numeric(df['Paladino'], errors='coerce')
df['Nobre'] = pd.to_numeric(df['Nobre'], errors='coerce')
df['Milicia'] = pd.to_numeric(df['Milicia'], errors='coerce')

df = df.fillna(0)
```

```python
df.convert_dtypes().dtypes
```

```python
df['Total Tropas Defesa'] = df['Lança'] + df['Espada'] + df['Cavalaria Pesada']
```

```python
df['Total Tropas Ataque'] = df['Machado'] + df['Cavalaria Leve'] + df['Ariete']
```

```python
df['Total Tropas'] = df['Total Tropas Defesa'] + df['Total Tropas Ataque'] + df['Cataputa']
```

```python
df['Porcentagem Tropas Ataque'] = (df['Total Tropas Ataque'] / df['Total Tropas']) * 100
```

```python
df['Porcentagem Tropas Defesa'] = (df['Total Tropas Defesa'] / df['Total Tropas']) * 100
```

```python
df['Tipo Aldeia'] = pd.cut(df['Porcentagem Tropas Ataque'], 3, labels=['Defesa', 'Misto', 'Ataque'])
```

```python
bins_bb = pd.IntervalIndex.from_tuples([(1000, 3000), (3001, 5000), (5000, 7000), (7001, np.inf)])
df['BB Qtd'] = pd.cut(df['Machado'].values, bins_bb, labels=["Poucas", "Meio Full", "Quase Full", "Full"])

bins_cl = pd.IntervalIndex.from_tuples([(0, 500), (501, 1500), (1501, 2500), (2501, np.inf)])
df['BB Qtd'] = pd.cut(df['Cavalaria Leve'].values, bins_cl, labels=["Poucas", "Meio Full", "Quase Full", "Full"])
```
